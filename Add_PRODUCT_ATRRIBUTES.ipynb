## Fuzzy matching implementation

### Subtask:
Implement a fuzzy matching function to compare product names in German, considering potential variations and typos.

Define the fuzzy matching function using the installed libraries.



```python
from fuzzywuzzy import fuzz

def german_fuzzy_match(name1, name2):
  """
  Calculates the fuzzy similarity score between two German product names.

  Args:
    name1: The first product name (string).
    name2: The second product name (string).

  Returns:
    An integer representing the similarity score (0 to 100).
  """
  return fuzz.token_sort_ratio(name1, name2)

# Example usage (optional, for testing)
# score = german_fuzzy_match("Apple iPhone 13", "iPhone 13 Apple")
# print(f"Similarity score: {score}")
```

## Apply fuzzy matching and join

### Subtask:
Apply the fuzzy matching function to find the best matches between the two DataFrames and perform the join based on these matches.


**Reasoning**:
Apply the fuzzy matching function to find the best matches between the two DataFrames and perform the join based on these matches.



```python
def find_best_match(product_name, products_df, threshold=80):
    """
    Finds the best matching Product_NUM from products_df for a given product name.

    Args:
        product_name: The product name from orders_df (string).
        products_df: The DataFrame containing product information (pandas DataFrame).
        threshold: The minimum fuzzy matching score to consider a match (integer, 0-100).

    Returns:
        The Product_NUM of the best match, or None if no match is found above the threshold.
    """
    best_score = -1
    best_match_product_num = None

    for index, row in products_df.iterrows():
        score = german_fuzzy_match(product_name, row['Product_Name'])
        if score > best_score and score >= threshold:
            best_score = score
            best_match_product_num = row['Product_NUM']

    return best_match_product_num

# Apply the function to orders_df to create a new column with matching Product_NUMs
orders_df['matched_Product_NUM'] = orders_df['productName'].apply(lambda x: find_best_match(x, products_df))

# Perform a left join of orders_df and products_df
joined_df = pd.merge(orders_df, products_df, left_on='matched_Product_NUM', right_on='Product_NUM', how='left')

# Display the head of the resulting joined DataFrame
display(joined_df.head())
```

## Create new table

### Subtask:
Write the resulting joined DataFrame to a new table named `Orders_Product_Enhanced` in the database.


**Reasoning**:
Write the resulting joined DataFrame to a new table named Orders_Product_Enhanced in the database.



```python
import sqlite3

db_path = '/content/Categorized_DB_Enhanced (2).db'
table_name = 'Orders_Product_Enhanced'

try:
    conn = sqlite3.connect(db_path)
    # Write the DataFrame to a new table in the database
    joined_df.to_sql(table_name, conn, if_exists='replace', index=False)

    print(f"DataFrame successfully written to table '{table_name}' in the database.")

except sqlite3.Error as e:
    print(f"Error writing DataFrame to database: {e}")

finally:
    if conn:
        conn.close()
```
